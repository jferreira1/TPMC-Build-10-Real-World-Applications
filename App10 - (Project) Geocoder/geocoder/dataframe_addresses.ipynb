```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import pandas as pd


df = pd.DataFrame(columns=["Enderecos", "Bairros"])
df.iloc[0:0]
```

```python
bairro = "centro"
cidade = "niteroi"
estado = "rj"
url = "https://paginaamarela.com.br/bairro/"+ bairro +"/"+cidade+"/"+estado+"#enderecos"
r = requests.get(url)
c = r.content
soup = BeautifulSoup(c, "html.parser")

underline = soup.find_all('a', {'class':'underline'})

address = []
for line in underline:
    if "cep" in line.text.lower():
        pass
    else:
        address.append(line.text)

cidade = address[0]
address = address[1:-4]   
bairros = len(address)*[bairro.capitalize()]
cidades = len(address)*[cidade]
estados = len(address)*[estado.upper()]

data = {'Enderecos': address,
        'Bairros': bairros,
        'Cidades': cidades,
        'Estados': estados
        }

df2 = pd.DataFrame(data)
df.append(df2)
address.clear()
bairros.clear()

```

```python
df = df.append(df2, ignore_index=True)
df
```
