```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

r = requests.get("https://paginaamarela.com.br/bairro/arsenal/sao-goncalo/rj#enderecos")
c = r.content
soup = BeautifulSoup(c, "html.parser")
```

```python
underline = soup.find_all('a', {'class':'underline'})
pcode = []
address = []

for line in underline:
    print(line.text)
    if "cep" in line.text.lower():
        pcode.append(line.text.strip('Cep '))
    else:
        address.append(line.text)
        
address = address[1:-4]
    
```

```python

address
```

```python
import pandas as pd

df = pd.DataFrame()
df.iloc[0:0]
```

```python
df['Address'] = address
df['CEP'] = pcode
```

```python
df.to_csv('addresses.csv')
```
